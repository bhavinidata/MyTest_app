```python
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float, ForeignKey, ForeignKeyConstraint

!pip install psycopg2
!pip install psycopg2-binary
```

```python
user = "postgres"
password = "changeme"
host = "localhost"
port = "5432" # 5432 for everyone else
db = "test"
uri = f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{db}"
uri
```

```python
# Create the connection engine
engine = create_engine(uri)
conn = engine.connect()
```

```python
Base = declarative_base()
class My_data(Base):
    __tablename__ = "my_test_table"
    myid = Column(Integer, primary_key=True)
    myname = Column(String(255))
    myname1 = Column(String(255))
```

```python
Base.metadata.drop_all(conn)
Base.metadata.create_all(conn)

demographic_df.to_sql(My_data.__tablename__, conn, index=False, if_exists="append")
```
