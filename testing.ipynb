```python
Base = declarative_base()
class Demographic_data(Base):
    __tablename__ = "demographic_data"
    state_fips_code = Column(Integer, primary_key=True)     
    chsi_state_name = Column(String(255))
    county_fips_code = Column(Integer, primary_key=True)
    chsi_county_name = Column(String(255))
    chsi_state_abbr = Column(String(255))
    strata_id_number =  Column(Integer)     
    strata_determining_factors = Column(String(255))
    number_counties =  Column(Integer)     
    population_size =  Column(Integer)     
    poverty = Column(Float)
    age_19_under = Column(Float)
    age_19_64 = Column(Float)
    age_65_84 = Column(Float)
    age_85_and_over = Column(Float)
    white = Column(Float)
    black = Column(Float)
    native_american = Column(Float)
    asian = Column(Float)
    hispanic = Column(Float)

class Riskfactor_data(Base):
    __tablename__ = "riskfactor_data"
    riskfactor_id = Column(Integer, primary_key=True)
    state_code = Column(Integer)     
    county_code = Column(Integer)
    no_exercise = Column(Float)
    few_fruit_veg = Column(Float)
    obesity = Column(Float)
    high_blood_pres = Column(Float)
    smoker = Column(Float)
    diabetes = Column(Float)
    uninsured = Column(Float)
    prim_care_phys_rate = Column(Float)
    dentist_rate = Column(Float)
    community_health_center_ind = Column(Float)
    __table_args__ = (ForeignKeyConstraint([county_code, state_code],
                                           [Demographic_data.county_fips_code, Demographic_data.state_fips_code]),
                      {})

```

```python
# Create the connection engine
engine = create_engine(uri)
conn = engine.connect()

# Base.metadata.drop_all(conn)
Base.metadata.create_all(conn)

demographic_df.to_sql(Demographic_data.__tablename__, conn, index=False, if_exists="append")

riskfactor_df.to_sql(Riskfactor_data.__tablename__, conn, index=False, if_exists="append")
```
